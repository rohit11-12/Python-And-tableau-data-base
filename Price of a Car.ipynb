# Outlier Removal

```python
sns.boxplot(data['Price'])
```

```python
sorted(data['Price'],reverse=True)
```

```python
data = data[~(data['Price']>=31000) & (data['Price']<=32500)]
```

```python
X = data.drop(['Model','Price','Id'],axis=1)
Y = data['Price']
```

```python
X
```

# Store Feature Matrix in X and Response(Target) In Vector Y

```python
from sklearn.model_selection import train_test_split
```

# Splitting The Dataset into The Training Set And Test Set

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=142)
```

```python
data.head()
```

# Import The Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

# Model Training

```python
ar = LinearRegression()
ar.fit(X_train,Y_train)

rf = RandomForestRegressor()
rf.fit(X_train,Y_train)

xgb = GradientBoostingRegressor()
xgb.fit(X_train,Y_train)

xg = XGBRegressor()
xg.fit(X_train,Y_train)
```

```python
Y_pred1 = ar.predict(X_test)
Y_pred2 = rf.predict(X_test)
Y_pred3 = xgb.predict(X_test)
Y_pred4 = xg.predict(X_test)
```

# Model Score

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(Y_test,Y_pred1)
score2 = metrics.r2_score(Y_test,Y_pred2)
score3 = metrics.r2_score(Y_test,Y_pred3)
score4 = metrics.r2_score(Y_test,Y_pred4)
```

```python
print(score1,score2,score3,score4)
```
