```python
import datetime as dt
Latest_Date=dt.datetime(2011,12,10)
```

```python

RFMScore =data.groupby('CustomerID').agg({'InvoiceDate': lambda x: (Latest_Date- x.max()).days,'InvoiceNo': lambda x: len(x), 'TotalAmount': lambda x: x.sum()})
RFMScore.rename(columns={'InvoiceDate':'Recency',
                         'InvoiceNo': 'Frequency',
                         'TotalAmount':'Monetary'}, inplace=True)
RFMScore.reset_index().head()
```

```python
quantiles= RFMScore.quantile(q= [0.25, 0.5, 0.75])
quantiles= quantiles.to_dict()
```

```python
quantiles
```

```python
def RScoring(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4
    
    
def FnMScoring(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1        
```

```python
RFMScore['R'] = RFMScore['Recency'].apply(RScoring, args=('Recency',quantiles,))
RFMScore['F'] = RFMScore['Frequency'].apply(FnMScoring, args=('Frequency',quantiles,))
RFMScore['M'] = RFMScore['Monetary'].apply(FnMScoring, args=('Monetary',quantiles,))
RFMScore.head()
```

```python
RFMScore['RFMSGroup'] = RFMScore.R.map(str) + RFMScore.F.map(str) + RFMScore.M.map(str)
RFMScore['RFMscore']= RFMScore[['R','F','M']].sum(axis=1)
RFMScore.head()
```

```python
Loyalty_Level = ['Platinum','Gold','Silver','Bronze']
Score_cuts = pd.qcut(RFMScore.RFMscore, q=4, labels =Loyalty_Level)
RFMScore['RFM_Loyalty_level']= Score_cuts.values
RFMScore.reset_index().head()
```
